# Description of the project:

   In this part of the class, we will curate a set of **CO2 measurements**, measured from cheap but inaccurate sensors, that have been deployed in the city of Zurich. The goal of the exercise is twofold: 

1. Learn how to deal with real world sensor timeseries data, and organize them efficiently using python dataframes.

2. Apply data science tools to model the measurements, and use the learned model to process them (e.g., detect outliers and drifts in the sensor measurements). 

The sensor network consists of 46 sites, located in different parts of the city. Each site contains three different sensors measuring (a) **CO2 concentration**, (b) **temperature**, and (c) **humidity**. Beside these measurements, we have the following additional information that can be used to process the measurements: 


1. The average **daily wind pattern** for the city of Zurich 

2. The **altitude** at which the CO2 sensor is located

3. A clustering of the city of Zurich in 17 different zones and the **city zone** in which the sensor belongs to. Some characteristic zones are industrial area, residential area, forest, glacier, lake, etc.

Given this knowledge, you are asked to define a machine learning algorithm that curates the data, by detecting potential outliers or drifts. The algorithm should be based on the fact that sensors in similar conditions are expected to have similar measurements.

## Prior knowledge:
    
The average value of the CO2 in a city is approximately 400 ppm. However, the exact measurement in each site depends on parameters such as the temperature, the humidity, the wind, the altitude, and the level of traffic around the site. For example, sensors positioned in high altitude (mountains) and exposed to stronger winds, are expected to have a much lower and uniform level of CO2 than sensors that are positioned in a business area with much higher traffic activity. Moreover, we know that there is a strong dependence of the CO2 measurements, on temperature and humidity.



## Goal:
    
  We would like to use the above knowledge in order to fit a robust regression model to the CO2 measurements. This model should be built by processing the measurements of the entire sensor network jointly, and it should take into account the dependence on the above mentioned parameters. Finally, the learned model will be used to curate the CO2 measurements by detecting possible outliers and drifts. 

**Note:**
Different machine learning approaches can be used to solve the problem. For simplicity, we provide a suggested approach, and the corresponding steps that can guide your solution.

## Question I.a. (12/15)

> Organize the CO2_sensor_measurements.csv and temperature_humidity.csv files into a single dataframe that contains as index the time instance (timestamp) of the measurements, and columns the location of the site (LocationName), the sensor ID (SensorUnit_ID), the CO2 measurement, the temperature, and the humidity. For each measurement (CO2, humidity, temperature), take the average over an interval of 30 min. If there are missing measurements, interpolate them linearly from measurements that are close by in time.

>Hints: the following methods could be useful:

>- [pandas.DataFrame.resample()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html)
    
>- [pandas.DataFrame.interpolate()](https://pandas.pydata.org/pandasdocs/stable/generated/pandas.DataFrame.interpolate.html)

>- [pandas.DataFrame.mean()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.mean.html)

>- [pandas.DataFrame.append()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.append.html)

````python
#change the timestamp column format to datetime format
CO2_measurement['timestamp'] = CO2_measurement['timestamp'].apply(pd.to_datetime)
zurich_temp_humidity['Timestamp'] = zurich_temp_humidity['Timestamp'].apply(pd.to_datetime)
````

````python
sensor_list = [item.split('.')[0] for item in zurich_temp_humidity.filter(like='temperature').columns.tolist()]
````

````python
'''as the schema of the 'zurich_temp_humidity' dataset is different form 'CO2_measurements' dataset,'
we need to change the temp humidity dataset using the for loop below. for each sensor we do do a interpolation to 
fill the missing values for temperature and humidity
'''
zurich_temp_humidity = zurich_temp_humidity.rename(columns={'Timestamp':'timestamp'}).set_index('timestamp')
DF_list = []
for sensor in sensor_list:                                                 
    tmp = zurich_temp_humidity.filter(like=sensor)
    tmp.columns = ['temperature', 'humidity']
    tmp = tmp.interpolate()
    tmp['SensorUnit_ID'] = int(sensor)
    tmp = tmp.reset_index()
    DF_list.append(tmp)
    
temp_humidity_DF = pd.concat(DF_list)
````

````python
temp_humidity_DF.head()
````

````python
# resampling to make the 30 minutes intervals
temp_humidity_DF = temp_humidity_DF.set_index("timestamp").groupby('SensorUnit_ID')\
.resample('30T').mean().drop('SensorUnit_ID', axis=1)
````

````python
temp_humidity_DF = temp_humidity_DF.reset_index()
````

````python
# resampling to make 30 mins intervals for CO2 measurments
DF_list = []
for sensor in sensor_list:
    tmp = CO2_measurement[CO2_measurement['SensorUnit_ID'] == int(sensor)]
    tmp = tmp.set_index("timestamp").resample('30T').mean().drop('SensorUnit_ID', axis=1).reset_index()
    tmp = tmp.interpolate()
    tmp['SensorUnit_ID'] = [int(sensor)] * len(tmp)
    DF_list.append(tmp)
co2_DF = pd.concat(DF_list)
````

````python
merged_DF = temp_humidity_DF.merge(co2_DF, on=['timestamp','SensorUnit_ID'])
````

````python
merged_DF = merged_DF.set_index("timestamp")
````

````python
merged_DF.shape
````

After merging, there are many NaNs in LocationName. However, we know that each sensor has been installed in particular location, so using the ```sensor_id2loc_name``` dictionary we made earlier, we are going to convert sensor_IDs to location names:

````python
merged_DF['LocationName'] = merged_DF['SensorUnit_ID'].map(lambda x: sensor_id2loc_name[x])
````

````python
# Re-ordering columns to resemble the sample given
merged_DF = merged_DF[['CO2', 'LocationName', 'SensorUnit_ID', 'temperature', 'humidity']]
````

## Question II.b. (7/20)

**Altitude clustering:** Cluster the sites based on their altitude, by performing a K-means clustering. Set the number of clusters to 4. 

````python
altitudes = np.array(sensors_metadata['altitude'])
altitudes = altitudes.reshape((len(altitudes), 1))
LocNames = np.array(sensors_metadata['LocationName'])
kmeans = KMeans(n_clusters=4, random_state=0, algorithm='full').fit(altitudes)
kmeans.labels_
````

We can see the clusters and the location names that belong to them

````python
cluster2map_loc = {u:[] for u in range(4)}
for i,c in enumerate(kmeans.labels_):
    cluster2map_loc[c].append(LocNames[i])
for k in range(4):
    print("cluster " + str(k) + ":", cluster2map_loc[k] , "\n")
````

````python
# Creating the inverse look-up map: cluster# <=> location
map_loc2cluster = {}
for k, v in cluster2map_loc.items():
    for v_ in v:
        map_loc2cluster[v_] = k
````

````python
# Adding alt_cluster# to the table
merged_DF['alt_cluster'] = merged_DF['LocationName'].apply(lambda x: map_loc2cluster[x])
````

> **Temporal clustering:** Cluster the days of the month based on the average wind conditions of the city of Zurich. 

> First, resample the daily wind signal by computing the mean over an interval of 30 min. 

> Then, consider each day as a 48-dimensional sample (30 min intervals per day). Perform a principle component analysis of the wind values to reduce the dimension to 2.  
Hint: You can use `pandas.DataFrame.grouby()` and `for` to generate the daily sample ([documentation](https://pandas.pydata.org/pandas-docs/stable/groupby.html#iterating-through-groups)).

````python
zurich_average_wind['timestamp'] = zurich_average_wind['timestamp'].apply(pd.to_datetime)
zurich_average_wind['day'] = zurich_average_wind['timestamp'].map(lambda x: x.day)
perday_zurich_average_wind = zurich_average_wind.set_index('timestamp')
````

````python
#aggregating measurments in each day in a list
perday_zurich_average_wind = perday_zurich_average_wind.groupby('day').aggregate(lambda x: list(x))
perday_zurich_average_wind.head()
````

````python
# Do the PCA
X = np.array(perday_zurich_average_wind['wind_speed'])
daily_wind=[]
for item in X:
    daily_wind.append(item)
daily_wind = np.array(daily_wind)
# daily_wind.shape
pca = PCA(n_components=2)
pca.fit(daily_wind)
````

````python
#apply dimentionality reduction to daily eind data
daily_wind = pca.transform(daily_wind)
````

````python
kmeans = KMeans(n_clusters=6, random_state=0, algorithm='full').fit(daily_wind)
kmeans.labels_
````

The days that belong to each cluster

````python
cluster2map_day = {u:[] for u in range(6)}
for i,c in enumerate(kmeans.labels_):
    cluster2map_day[c].append(i+1)
for k in range(6):
    print("cluster " + str(k) + ":", cluster2map_day[k] , "\n")
````

````python
# Creating the inverse look-up map: day# <=> cluster
map_day2cluster = {}
for k, v in cluster2map_day.items():
    for v_ in v:
        map_day2cluster[v_] = k
````

````python
# Adding the day# to merged_DF
merged_DF['day'] = merged_DF.index.to_series().apply(lambda x: x.day)
````

````python
# Adding wind_cluster# to the table
merged_DF['wind_cluster'] = merged_DF['day'].apply(lambda x: map_day2cluster[x])
````

We next do the partitioning and training of the classifiers:

````python
# Partitioning the DF into 6 wind clusters

clustered_data = {_c:{'dates':None, 'features':None, 'targets':None} for _c in range(6)}

for _c in range(6):
    clustered_data[_c]['dates'] = filtered_merged_DF.loc[filtered_merged_DF['wind_cluster'] == _c].sort_index().index.to_series().values
    clustered_data[_c]['features'] = filtered_merged_DF.loc[filtered_merged_DF['wind_cluster'] == _c].sort_index()[['humidity', 
                                                                                                       'temperature',
                                                                                                       'ts_id']].values
    clustered_data[_c]['targets'] = filtered_merged_DF.loc[filtered_merged_DF['wind_cluster'] == _c].sort_index()['CO2'].values
````

````python
# Training a set of classifiers, one per each wind cluster
clf_set = []

for _c in range(6):
    _lr = linear_model.LinearRegression()
    clf_set.append(_lr.fit(clustered_data[_c]['features'], clustered_data[_c]['targets']))
````
